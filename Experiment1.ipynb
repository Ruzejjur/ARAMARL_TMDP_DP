```python
import numpy as np

from engine import RMG
from agent import RandomAgent, IndQLearningAgent
```

```python
N_EXP = 10

r0ss = []
r1ss = []
```

```python
for n in range(N_EXP):
    batch_size = 1
    max_steps = 20
    gamma = 0.96

    # Reward matrix for the Iterated Prisoner's Dilemma
    ipd_rewards = np.array([[-1., 0.], [-3., -2.]])


    env = RMG(max_steps=max_steps, payouts=ipd_rewards, batch_size=batch_size)
    env.reset()

    possible_actions = [0, 1]  # Cooperate or Defect

    #cooperator, defector = RandomAgent(possible_actions, p=0.9), RandomAgent(possible_actions, p=0.1)
    p0, p1 = IndQLearningAgent(possible_actions, n_states=1, learning_rate=0.1, epsilon=0.05, gamma=gamma),\
        IndQLearningAgent(possible_actions, n_states=1, learning_rate=0.1, epsilon=0.05, gamma=gamma)


    #cooperator, defector = IndQLearningAgent(possible_actions, n_states=1, learning_rate=0.05, epsilon=0.1, gamma=gamma), \
    #    IndQLearningAgent(possible_actions, n_states=1, learning_rate=0.05, epsilon=0.1, gamma=gamma)


    # Stateless interactions (agents do not have memory)
    s = 0

    n_iter = 1000

    r0s = []
    r1s = []

    for i in range(n_iter):

        # A full episode:
        done = False

        while not done:

            # Agents decide
            a0 = p0.act()
            a1 = p1.act()

            # World changes
            _, (r0, r1), done, _ = env.step(([a0], [a1]))

            # Agents learn

            p0.update(s, (a0, a1), (r0, r1), s )
            p1.update(s, (a1, a0), (r1, r0), s )

            #s = new_s  #stateless!

            #print(r0, r1)
            r0s.append(r0[0])
            r1s.append(r1[0])

        env.reset()
    
    print(n)
    r0ss.append(r0s)
    r1ss.append(r1s)
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

We report moving avearage of rewards, since it's common in RL taks

```python
%matplotlib inline
import matplotlib.pyplot as plt

# We set a fancy theme
plt.style.use('ggplot')

plt.axis([0, max_steps*n_iter, -3.5, 0.5])

plt.plot(moving_average(r0s, 100), 'c')
plt.plot(moving_average(r1s, 100))
plt.xlabel('t');
plt.ylabel('R');

plt.savefig('img/QvsQ.png')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# We set a fancy theme
plt.style.use('ggplot')

plt.axis([0, max_steps*n_iter, -3.5, 0.5])
for i in range(N_EXP):
    plt.plot(moving_average(r0ss[i], 100), 'b', alpha=0.05)
    plt.plot(moving_average(r1ss[i], 100), 'r', alpha=0.05)
    
plt.plot(moving_average(np.asarray(r0ss).mean(axis=0), 100), 'b', alpha=0.5)
plt.plot(moving_average(np.asarray(r1ss).mean(axis=0), 100), 'r', alpha=0.5)

plt.xlabel('t');
plt.ylabel('R');

plt.legend(['Agent A', 'Agent B']);

plt.savefig('img/QvsQ.png')
```
